```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('expert180_samples.csv')
df
```

```python
# Create a new figure with a specified size (15 units in width and 8 units in height)
plt.figure(figsize=(15, 8))

# Create a violin plot using Seaborn
# x: Data in the 'True-label' column will be on the x-axis
# y: Data in the 'Confidence_score' column will be on the y-axis
# hue: The 'Expert_pred' column will be used to distinguish data points with different colors
# data:  DataFrame 'df'
# palette: 'muted' is the color palette to use for the plot
# split: Set to True, which means the violins will be split for each hue level
# inner: Show quartiles within the violins
# linewidth: Set the width of the line around the violins
sns.violinplot(x='True-label', y='Confidence_score', hue='Expert_pred', data=df, palette='muted', split=True, inner='quart', linewidth=1)

# Set the label for the x-axis
plt.xlabel('True Label')

# Set the label for the y-axis
plt.ylabel('Confidence Score')

# Set the title for the plot
plt.title('True Label vs. Confidence Score with Expert Predictions')

# Display the plot
plt.show()
```

Findings:
1. Both class 0 and class 1 are wrongly predicted
2. class 0 samples have high confidence score for correct predictions and meanwhile for class 1
3. class 0's are wrongly predicted but at low confidence scores like 4 and 3.
4. Most class 1's which are wrongly predicted are at low confidence scores.
5. class 1 at higer confidence score are even denser shows most class 1's are correctly predicted.
